```python
filepath = "sub-01_ses-20140827T193800_behavior+ecephys.nwb"
# Open the file in read mode "r",
io = NWBHDF5IO(filepath, mode="r", load_namespaces=True)
nwbfile = io.read()
nwbfile
```

```python
#Obtain the Participant Response and the Response Time
import pandas as pd
behavior_module = nwbfile.processing['behavior']                # Get 'behavior' processing module
behavior_iface  = behavior_module.data_interfaces['BehavioralEvents.response']
response_ts     = behavior_iface.time_series['response']       # The TimeSeries object

# Extract the data and timestamps as NumPy arrays
response_data     = response_ts.data[:]         # shape (N,)
response_timestamps = response_ts.timestamps[:] # shape (N,)

# (Optional) If responses are stored as bytes/strings, convert them
# e.g., decode to string or parse as boolean. This depends on your dataset.
# For example, if stored as bytes that represent 'Yes'/'No':
# response_data = [v.decode('utf-8') for v in response_data]

# Or if they are 0/1 or "True"/"False", cast them to actual booleans
# response_data = response_data.astype(bool)

# Create a DataFrame
df = pd.DataFrame({
    'timestamp': response_timestamps,
    'response': response_data
})

df.head()  # show first few rows
```

```python
#Obtain Trials information
trials_df = nwbfile.trials.to_dataframe()
merged_df = pd.concat([df, trials_df], axis=1)
merged_df.drop('timeseries', axis=1, inplace=True)
merged_df['Accuracy'] = np.where(merged_df['response'] == merged_df['solution'], 'Correct', 'Incorrect')
merged_df.index = merged_df.index + 1
```

```python
units_df = units_df.explode('electrodes')

# 4.2. Rename the column for clarity (optional)
units_df = units_df.rename(columns={'electrodes': 'electrode_idx'})

# 4.3. Merge with the electrodes_df on electrode index
units_merged_df = pd.merge(
    units_df,
    electrodes_df, 
    left_on='electrode_idx',   # the electrode index references
    right_index=True,          # the index of electrodes_df is the electrode ID
    how='left'
)

units_merged_df.head()
```

```python
#Obtain Spike data
spikes_df= nwbfile.units.to_dataframe()
locations=[];
#Identify which region of the brain is being recorded
for i in range(spikes_df['electrodes'].shape[0]):
    first_electrode_group = spikes_df['electrodes'].iloc[i]
    first_location = first_electrode_group.label.values[0]
    locations.append(first_location)
neuron_locs = pd.DataFrame(locations, columns=['Brain Region'])
neuron_locs.index = neuron_locs.index + 1
spikes_rev = pd.concat([spikes_df, neuron_locs], axis=1)
spikes_rev
```

```python
#33 Neurons in spike_df
#Each Neuron has spike times for all trials, as well as location
#For each neuron, want to separate firing by trial
#Trial information is in merged_df
# 2. Prepare column names for the new spike-time columns
neuron_columns = []
for idx, row in spikes_rev.iterrows():
    # Use index+1 to get numbering starting at 1
    col_name = f"neuron{idx+1}[{row['Brain Region']}]"
    neuron_columns.append(col_name)
# 3. Create an empty DataFrame with those columns, matching merged_df index
trial_spike_times = pd.DataFrame(index=merged_df.index, columns=neuron_columns)

# 4. Fill it with filtered spike times
for trial_idx, trial_row in merged_df.iterrows():
    start_t = trial_row['start_time']
    stop_t = trial_row['stop_time']
    
    # Loop over each neuron.
    for idx, neuron_row in spikes_rev.iterrows():
        # Create a column name using the neuron index and brain region.
        col_name = f"neuron{idx+1}[{neuron_row['Brain Region']}]"
        # Get the neuron's spike times.
        spikes = neuron_row['spike_times']
        # Filter spike times that are between start_t and stop_t.
        filtered_spikes = [t for t in spikes if start_t <= t <= stop_t]
        # Store the filtered list in the corresponding cell.
        trial_spike_times.loc[trial_idx, col_name] = filtered_spikes
# 5. Concatenate with merged_df
final_df = pd.concat([merged_df, trial_spike_times], axis=1)

# final_df now has 189 rows (trials) and 33 additional columns, each cell being a list of spike times.
final_df
```

```python
#Isolating a single set_size
filtered_df = final_df[final_df['set_size'] == 4].copy()
filt_dfv2=filtered_df.drop("solution", axis=1)
filt_dfv2
```
