# Compute Firing Rate Bins

```python

def compute_firing_rate(spike_times, start_time, stop_time, bin_width=0.1):
    """
    Compute binned firing rate for one trial.
    
    Parameters
    ----------
    spike_times : array-like
        Times (in seconds) of all spikes in this trial (must lie between start and stop).
    start_time : float
        Trial start time (in seconds).
    stop_time : float
        Trial end time (in seconds).
    bin_width : float, optional
        Width of each time bin in seconds (default: 0.1s).
    
    Returns
    -------
    rates_df : pandas.DataFrame
        DataFrame with columns:
          'time_bin_center' : center of each bin
          'firing_rate'     : spikes/sec in that bin
    """
    # 1) Define the bin edges
    edges = np.arange(start_time, stop_time + bin_width, bin_width)
    
    # 2) Count spikes in each bin
    counts, _ = np.histogram(spike_times, bins=edges)
    
    # 3) Convert to rate (spikes per second)
    rates = counts / bin_width
    
    # 4) Compute bin centers for plotting/analysis
    centers = edges[:-1] + bin_width/2.0
    
    # 5) Return a nice DataFrame
    rates_df = pd.DataFrame({
        'time_bin_center': centers,
        'firing_rate':    rates
    })
    return rates_df

# --- Example usage ---
# Suppose your trial starts at 1.0s, ends at 2.5s, and these are the spike times within that window:
spikes = filtered_df4f.iloc[1]['neuron3[AHL2]']
start, stop = filtered_df4f.iloc[1]['start_time'], filtered_df4f.iloc[1]['stop_time']
bin_w = 0.2   # 200ms bins

fr_df = compute_firing_rate(spikes, start, stop, bin_w)
print(fr_df)
```

```python
import matplotlib.pyplot as plt

# Extract your x and yvalues
x = fr_df['time_bin_center']
y = fr_df['firing_rate']

# Compute bin width (assumes uniform spacing)
bin_width = x.iloc[1] - x.iloc[0]

# Plot
plt.figure()
plt.bar(x, y, width=bin_width, align='center')
plt.xlabel('Time (s)')
plt.ylabel('Firing rate (Hz)')
plt.title('Binned Firing Rate')
plt.show()
```

# Cost

```python
spike_times_list = filtered_df4f['neuron3[AHL2]'].tolist()   # length N_trials
start_time = filtered_df4f['start_time'].min()               # assume same window
stop_time  = filtered_df4f['stop_time'].max()
n_trials   = len(spike_times_list)

def cost_function(delta):
    """
    Compute C_n = (2 * k_bar - v) / ((n_trials * delta)**2)
    where k_bar = mean spikecount per bin (averaged over all bins & trials)
          v     = variance of those bincounts
    """
    # define bin edges once
    edges = np.arange(start_time, stop_time + delta, delta)

    # build a (n_trials  n_bins) array of counts
    counts = np.vstack([
        np.histogram(spikes, bins=edges)[0]
        for spikes in spike_times_list
    ])  # shape = (n_trials, n_bins)

    # flatten to 1D if you like, or compute perbin across trials:
    all_counts = counts.ravel()

    k_bar = all_counts.mean()
    v     = all_counts.var(ddof=0)

    return (2 * k_bar - v) / ( (n_trials * delta)**2 )


# 2) Sweep over candidate binwidths and find the mincost (i.e. best) 
deltas = np.linspace(0.01, 1.0, 100)   # e.g. 10ms1s in 100 steps
costs  = np.array([cost_function(d) for d in deltas])

best_idx    = np.argmin(costs)
best_delta  = deltas[best_idx]
best_cost   = costs[best_idx]
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(deltas, costs)
plt.xlabel('Bin width  (s)')
plt.ylabel('Cost C()')
plt.title('Cost vs. Bin Width for neuron3[AHL2]')
plt.show()
```
